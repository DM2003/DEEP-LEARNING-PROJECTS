```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
# Adding target column to dataframe (1 for benign and 0 for malignant)
data_frame['label'] = breast_cancer_dataset.target
```

```python
# Separating features and target
X = data_frame.drop(columns = 'label', axis = 1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Standardizing the data
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train) # To bring all values in more or less of the same range
X_test_std = scaler.transform(X_test)

# No need to transform Y_train and Y_test since it's already in form of 0 and 1.
```

```python
print(X_train_std)
```

**BUILDING THE NEURAL NETWORK**

```python
import tensorflow as tf
tf.random.set_seed(3) # We use this so that no matter how many times we run our code we get the same accuracy. If we don't use this, accuracy may change each time we run our code.
from tensorflow import keras
```

```python
# Setting up the layers of my neural network
model = keras.Sequential([
                         keras.layers.Flatten(input_shape = (30,)), # Thus we have one input layer. Flatten will convert the entire matrix of X values to a single dimensional array. Inside input_shape mention the no of columns that we have WITHOUT THE TARGET VARIABLE.
                         keras.layers.Dense(20, activation = 'relu'), # Here we will have one hidden layer. Dense indicates that all the neurons in that layer is connected to all neurons in previous layer. 20 is the no of neurons we want in our hidden layer and we are using the Relu activation function.
                         keras.layers.Dense(2, activation = 'sigmoid') # Here we will have one input layer

])
```

```python
# Compiling the neural network
model.compile(optimizer = 'adam',
              loss = 'sparse_categorical_crossentropy', # Whenever we have this kind of dataset where target variables are represented directly as 0 or 1 or 2 and so on, we will use this loss function.
              metrics = ['accuracy'])

# Whenever we have one not encoding kind of dataset we will use categorical_crossentropy.
# One not encoding dataset:
# Suppose, benign is 0 and malignant is 1 and instead of denoting as 0 or 1 we will represent as:
# Benign = 001 (The last bit is 1)
# Malignant = 010 (The second last bit is 1) and so on
```

```python
# Training our model
history = model.fit(X_train_std, Y_train, validation_split = 0.1, epochs = 10) # Epochs is how many times model will go through the data
```

```python
# Visualizing the accuracy and loss
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title("Model accuracy")
plt.ylabel("Accuracy")
plt.xlabel("Epochs")

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title("Model Loss")
plt.xlabel(" Loss")
plt.ylabel("Epochs")

plt.legend(['training data', 'validation data'], loc = 'lower left')
```

```python
# Accuracy of the model on test data
loss, accuracy = model.evaluate(X_test_std, Y_test)
print("Accuracy:", accuracy)
```
