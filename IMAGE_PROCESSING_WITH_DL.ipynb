** Libraries that can be used for image processing **
1. matplotlib.image
2. Pillow
3. OpenCV(cv2)

```python
# Importing the image module from matplotlib library
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
# Loading an image through matplotlib.image module
img = mpimg.imread('/content/dog.jpg')
```

```python
type(img)
```

Thus, we can see our jpg image has been converted to numpy array

```python
print(img.shape)
```

The first and second value represents the height and width of the image respectively and the third value represents number of color channels in the image. Here, we had an RGB image, so the color channel is 3 ( Red, Green and Blue )

```python
print(img)
```

```python
# Another thing we can do is that we can convert this numpy array to our original image
img_plot = plt.imshow(img)
plt.show()
```

```python
# Resizing the image using Pillow library
from PIL import Image
```

```python
img = Image.open('/content/dog.jpg')
img_resized = img.resize((100,100))
```

```python
img_resized.save('dog_image_resized.jpg')
```

```python
img_resized = mpimg.imread('/content/dog_image_resized.jpg')
img_resized_plot = plt.imshow(img_resized)
plt.show()
```

```python
print(img_resized.shape)
```

** Converting RGB images to Grayscale images using OpenCV **

```python
import cv2 # CV stands for Computer Vision
```

```python
img = cv2.imread('/content/dog.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
greyscale_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(greyscale_img)
```

```python
greyscale_img.shape
```

cv2_imshow() won't be allowed directly over here, hence we will use the following code:

```python
from google.colab.patches import cv2_imshow
```

```python
# Displaying the image
cv2_imshow(greyscale_img)
```

```python
# Saving the grayscale image
cv2.imwrite('Dog_greyscale_image.jpg', greyscale_img)
```
